```python
import pandas as pd
import numpy as np

# Read the parquet file
try:
    # Load the data
    df = pd.read_parquet('yellow_tripdata_2023-02.parquet')
    
    # Calculate the total number of records in the original dataset
    total_records = len(df)
    print(f"Total records in original dataset: {total_records:,}")
    
    # Compute trip duration in minutes
    df['trip_duration_minutes'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
    
    # Analyze the distribution of trip durations before filtering
    print("\nTrip Duration Distribution (before filtering):")
    print(f"Min: {df['trip_duration_minutes'].min():.2f} minutes")
    print(f"Max: {df['trip_duration_minutes'].max():.2f} minutes")
    print(f"Mean: {df['trip_duration_minutes'].mean():.2f} minutes")
    print(f"Median: {df['trip_duration_minutes'].median():.2f} minutes")
    print(f"Standard Deviation: {df['trip_duration_minutes'].std():.2f} minutes")
    
    # Show percentiles to understand the distribution
    percentiles = [1, 5, 10, 25, 50, 75, 90, 95, 99]
    print("\nPercentiles before filtering:")
    for p in percentiles:
        print(f"{p}th percentile: {df['trip_duration_minutes'].quantile(p/100):.2f} minutes")
    
    # Count records outside our target range (1-60 minutes)
    less_than_1min = (df['trip_duration_minutes'] < 1).sum()
    more_than_60min = (df['trip_duration_minutes'] > 60).sum()
    
    print(f"\nRecords with duration < 1 minute: {less_than_1min:,} ({less_than_1min/total_records:.2%})")
    print(f"Records with duration > 60 minutes: {more_than_60min:,} ({more_than_60min/total_records:.2%})")
    
    # Filter to keep only records with duration between 1 and 60 minutes (inclusive)
    filtered_df = df[(df['trip_duration_minutes'] >= 1) & (df['trip_duration_minutes'] <= 60)]
    
    # Calculate the number of records after filtering
    filtered_records = len(filtered_df)
    
    # Calculate the fraction of records remaining
    fraction_remaining = filtered_records / total_records
    
    print(f"\nRecords after filtering (1-60 minutes): {filtered_records:,}")
    print(f"Fraction of records remaining: {fraction_remaining:.6f} ({fraction_remaining:.2%})")
    
    # Show distribution statistics after filtering
    print("\nTrip Duration Distribution (after filtering):")
    print(f"Min: {filtered_df['trip_duration_minutes'].min():.2f} minutes")
    print(f"Max: {filtered_df['trip_duration_minutes'].max():.2f} minutes")
    print(f"Mean: {filtered_df['trip_duration_minutes'].mean():.2f} minutes")
    print(f"Median: {filtered_df['trip_duration_minutes'].median():.2f} minutes")
    print(f"Standard Deviation: {filtered_df['trip_duration_minutes'].std():.2f} minutes")
    
except FileNotFoundError:
    print("Error: The file 'yellow_tripdata_2023-02.parquet' was not found.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer

try:
    # Load the parquet file
    df = pd.read_parquet('yellow_tripdata_2023-02.parquet')
    
    # Calculate trip duration in minutes
    df['trip_duration_minutes'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60
    
    # Filter to keep only records with duration between 1 and 60 minutes (inclusive)
    filtered_df = df[(df['trip_duration_minutes'] >= 1) & (df['trip_duration_minutes'] <= 60)]
    
    print(f"Records after filtering (1-60 minutes): {len(filtered_df):,}")
    
    # Check the column names to identify pickup and dropoff location IDs
    print("\nAvailable columns:")
    print(filtered_df.columns.tolist())
    
    # Assuming the location IDs are 'PULocationID' and 'DOLocationID'
    # Extract only the needed columns
    features_df = filtered_df[['PULocationID', 'DOLocationID']]
    
    # Check for missing values
    missing_values = features_df.isnull().sum()
    print(f"\nMissing values:\n{missing_values}")
    
    # Drop rows with missing values if any
    if missing_values.sum() > 0:
        features_df = features_df.dropna()
        print(f"Dropped rows with missing values. Remaining rows: {len(features_df):,}")
    
    # Convert location IDs to strings to ensure proper one-hot encoding
    features_df['PULocationID'] = 'pickup_' + features_df['PULocationID'].astype(str)
    features_df['DOLocationID'] = 'dropoff_' + features_df['DOLocationID'].astype(str)
    
    # Convert DataFrame to list of dictionaries
    dict_records = features_df.to_dict(orient='records')
    
    print(f"\nSample of dictionary records (first 3):")
    for i, record in enumerate(dict_records[:3]):
        print(f"{i+1}: {record}")
    
    # Initialize and fit the DictVectorizer
    dv = DictVectorizer(sparse=False)
    X = dv.fit_transform(dict_records)
    
    # Get the feature names
    feature_names = dv.get_feature_names_out()
    
    # Calculate dimensionality (number of columns)
    dimensionality = X.shape[1]
    
    print(f"\nDimensionality of the feature matrix: {dimensionality}")
    print(f"Shape of the feature matrix: {X.shape}")
    
    # Count unique pickup and dropoff locations
    unique_pickups = len([f for f in feature_names if f.startswith('pickup_')])
    unique_dropoffs = len([f for f in feature_names if f.startswith('dropoff_')])
    
    print(f"\nNumber of unique pickup locations: {unique_pickups}")
    print(f"Number of unique dropoff locations: {unique_dropoffs}")
    print(f"Total unique locations: {len(set([int(f.split('_')[1]) for f in feature_names]))}")
    
    # Show a sample of the one-hot encoded matrix
    print("\nSample of the feature matrix (first 3 rows, first 10 columns):")
    print(X[:3, :10])
    
except FileNotFoundError:
    print("Error: The file 'yellow_tripdata_2023-02.parquet' was not found.")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```
